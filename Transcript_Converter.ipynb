```python
import json
```

```python
def split_transcript(transcript):
  split = transcript.split('\n\n')
  arr = []
  for item in split:
    format = item.split(': ', 1)
    format = json.dumps({'speaker': format[0], 'utterance': format[1]})
    arr.append(format)
  print(arr)
  return arr
```

```python
with open('transcript.txt', 'r') as f:
  data = f.read()
  text = split_transcript(data)
```
